```python
#preprocessing converting categorical columns into continuos
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
A=A.apply(le.fit_transform)
```

#Random forest

```python
#splitting dataset into training testing set for analysis
X=A.drop(labels=['MonthlyRate'], axis=1)
Y=A[['MonthlyRate']]

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(X,Y, test_size=0.20, random_state=40)
```

```python
#using ensemble learning algorithm
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=10,random_state=40)
model=rfr.fit(xtrain, ytrain)
pred=model.predict(xtest)

from sklearn.metrics import mean_absolute_error, mean_squared_error
print('mae, is %.2f'%(mean_absolute_error(ytest, pred)))
print('mse, is %.2f'%(mean_squared_error(ytest, pred)))
```

```python
#Adaboost algorithm
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
abc=AdaBoostRegressor(DecisionTreeRegressor(max_depth=2), n_estimators=20)
model=abc.fit(xtrain, ytrain)
pred=model.predict(xtest)

from sklearn.metrics import mean_absolute_error
mean_absolute_error(ytest, pred)
```

#linear regression with ols technique

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain, ytrain)
pred = model.predict(xtest)

from sklearn.metrics import mean_absolute_error
mean_absolute_error(ytest, pred)
```
